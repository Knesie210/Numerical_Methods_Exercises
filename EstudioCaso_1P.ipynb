## TEORÍA DE ALGEBRA LINEAL PARA SISTEMAS DE ECUACIONES LINEALES

Para comprender Álgebra Lineal aplicado a Sistemas de Ecuaciones lineal se deben considerar los siguientes conceptos fundamentales:

\\
##1. Sistema de ecuaciones lineales:
Un sistema de ecuaciones lineales es un conjunto de ecuaciones de la forma:

$\large \begin{matrix}
a_{11} x_{1} & +a_{12} x_{2} & +.... & +a_{1n} x_{n} & =b_{1}\\
a_{21} x_{1} & +a_{22} x_{2} & +.... & +a_{2n} x_{n} & =b_{2}\\
 &  &  & . & \\
 &  &  & . & \\
 &  &  & . & \\
a_{m1} x_{1} & +a_{m2} x_{2} & +.... & +a_{mn} x_{n} & =b_{n}
\end{matrix}$

\\
##2. Matriz de coeficientes y vector de constantes

Un sistema de ecuaciones lineales se puede representar de manera matricial como Ax = B, donde:

$\large A=\begin{pmatrix}
a_{11} & a_{12} & ... & a_{1n}\\
a_{21} & a_{12} & ... & a_{2n}\\
. & . & . & .\\
. & . & . & .\\
. & . & . & .\\
a_{m1} & a_{m2} & ... & a_{mn}
\end{pmatrix} ,\ X=\begin{pmatrix}
x_{1}\\
x_{2}\\
x_{3}
\end{pmatrix} ,\ B=\begin{pmatrix}
b_{1}\\
b_{1}\\
b_{m}
\end{pmatrix}$

\\
Conformados por:

La matriz de cofactores A

La matriz de vectores de incógnitas X

La matriz de vectores de resultados B


\\
##3. Métodos de resolución

Algunos métodos comunes para resolver sistemas de ecuaciones lineales incluyen:



*   **Método de eliminación de Gauss**: Consiste en transformar el sistema en una forma triangular para resolver por sustitución hacia atrás.

*   **Inversa de la matriz**: Si A es invertible, se puede encontrar la solución como: $\large x=A^{-1} B$

*   **Método de eliminación de Gauss-Jordan**: Extiende el método de Gauss llevando la matriz a su forma reducida por filas

*   **Método de Cramer - determinantes:** Utiliza determinantes para resolver sistemas de* n* ecuaciones con *n* incognitas, siempre que el determinante de la matriz de coeficientes sea no nulo; para la resolución del mismo resulta excesivamente costosa: computacionalmente, es ineficiente para grandes matrices y por ello no es usado en aplicaciones prácticas que pueden implicar muchas ecuaciones.
*   **Método de sustitución regresiva con eliminación de Gaussiana**:  se resuelve un sistema lineal con una matriz
triangular inferior L.
*   **Método de eliminación Gaussiana** Se transforma el sistema original en un sistema equivalente en forma triangular superior, de manera que se puede resolver fácilmente mediante sustitución hacia atrás.
*   **Técnicas de Pivoteo**:
      - **Sin Pivoteo** : Se refiere cuando los pivotes utilizados para la eliminación de Gauss son los elementos de la matriz que originalmente están en la diagonal.

      -** Pivoteo parcial**:  técnica utilizada que permite mejorar la estabilidad numérica del algoritmo. En cada paso de la eliminación, se selecciona el elemento de mayor valor absoluto en la columna actual (dentro de las filas restantes) como el pivote y, si es necesario, se intercambian filas para colocar este valor en la posición del pivote. Esto ayuda a minimizar los errores numéricos que pueden acumularse debido a la división por números muy pequeños.

      - **Pivoteo parcial escalado**: Esta técnica toma en cuenta no solo el valor absoluto de los elementos en la columna actual, sino también su relación con los demás elementos de la misma fila. Esto ayuda a evitar problemas numéricos que pueden surgir debido a la diferencia en magnitud de los coeficientes.

      - **Pivoteo total o completo**:técnica en el método de eliminación gaussiana que, en cada paso del algoritmo, selecciona el elemento de mayor valor absoluto en la submatriz restante (no solo en la columna actual) y realiza intercambios tanto de filas como de columnas para colocar este valor en la posición del pivote

##IMPLEMENTACIÓN DE LOS MÉTODOS NUMÉRICOS PARA RESOLVER EL SISTEMA DE ECUACIONES LINEALES (ELIMINACIÓN GAUSSIANA Y TÉCNICAS DE PIVOTEO)

###ALGORITMO DE ELIMINACIÓN GAUSSIANA

```python
# Definir la matriz aumentada
# La matriz A incluye tanto los coeficientes del sistema de ecuaciones lineales como los términos independientes en la última columna
A = np.array([
    [0.04, 0.01, 0.01, 0.06],  # Primera ecuación: 0.04x1 + 0.01x2 + 0.01x3 = 0.06
    [0.2, 0.5, 0.2, 0.3],      # Segunda ecuación: 0.2x1 + 0.5x2 - 0.2x3 = 0.3
    [1, 2, 4, 11]             # Tercera ecuación: 1x1 + 2x2 + 4x3 = 11
], dtype=float)

# Número de filas
# Aquí se obtiene el número de filas de la matriz A, que corresponde al número de ecuaciones
n = A.shape[0]

# Eliminación Gaussiana
# Este bucle convierte la matriz A en una matriz triangular superior
for i in range(n):
    # Hacer el pivote igual a 1
    # Divide toda la fila i por el elemento A[i, i] para hacer que el pivote (elemento diagonal) sea 1
    A[i] = A[i] / A[i, i]

    # Hacer ceros abajo del pivote
    # Este bucle hace que todos los elementos debajo del pivote en la columna i sean cero
    for j in range(i+1, n):
        # Resta de la fila j la fila i multiplicada por A[j, i] para eliminar el término A[j, i]
        A[j] = A[j] - A[j, i] * A[i]

# Sustitución hacia atrás
# Este bucle resuelve el sistema de ecuaciones mediante sustitución hacia atrás
x = np.zeros(n)  # Inicializa el vector de soluciones con ceros
for i in range(n-1, -1, -1):
    # Calcula el valor de la variable x[i]
    # Restando la suma de los productos de los elementos de la fila i y las soluciones encontradas previamente
    x[i] = A[i, -1] - np.sum(A[i, i+1:n] * x[i+1:n])

# Imprime la solución del sistema
print("La solución del sistema es:", x)
```

####**Algoritmo de eliminación Gaussiana: Pivoteo Parcial**

```python
def gaussian_elimination_partial(A):
    """
    Realiza la eliminación gaussiana con pivoteo parcial en la matriz A.

    Parámetros:
    A (numpy array): Matriz aumentada del sistema de ecuaciones lineales.

    Retorna:
    tuple: Una tupla que contiene la solución del sistema, el número de operaciones realizadas y el tiempo de ejecución en segundos.
    """

    n = A.shape[0]  # Número de filas en la matriz A
    start_time = time.time()  # Guardar el tiempo de inicio
    operations = 0  # Inicializar contador de operaciones

    # Eliminación Gaussiana con pivoteo parcial
    for i in range(n):
        # Encontrar el índice del máximo en valor absoluto en la columna i desde la fila i hasta la última
        max_row = np.argmax(np.abs(A[i:n, i])) + i

        # Intercambiar filas i y max_row
        A[[i, max_row]] = A[[max_row, i]]
        operations += 2 * n  # Contar operaciones de intercambio de filas

        # Normalizar la fila i para hacer el pivote igual a 1
        pivot = A[i, i]
        A[i] /= pivot
        operations += n  # Contar operaciones de la normalización

        # Eliminar elementos debajo del pivote en la columna i
        for j in range(i + 1, n):
            factor = A[j, i]
            A[j] -= factor * A[i]
            operations += n  # Contar operaciones de la eliminación

    # Sustitución hacia atrás
    x = np.zeros(n)  # Inicializar vector de solución
    for i in range(n - 1, -1, -1):
        # Calcular el valor de x[i] usando los valores ya calculados de x[i+1] a x[n-1]
        x[i] = A[i, -1] - np.sum(A[i, i + 1:n] * x[i + 1:n])
        operations += n  # Contar operaciones de la sustitución

    end_time = time.time()  # Guardar el tiempo de finalización

    # Devolver la solución, el número de operaciones y el tiempo de ejecución
    return x, operations, end_time - start_time

# Ejemplo de uso con la misma matriz A
A = np.array([
    [0.04, 0.01, 0.01, 0.06],  # Primera ecuación: 0.04x1 + 0.01x2 + 0.01x3 = 0.06
    [0.2, 0.5, 0.2, 0.3],      # Segunda ecuación: 0.2x1 + 0.5x2 - 0.2x3 = 0.3
    [1, 2, 4, 11]              # Tercera ecuación: 1x1 + 2x2 + 4x3 = 11
], dtype=float)

# Llamar a la función y obtener resultados
x_partial, operations_partial, time_partial = gaussian_elimination_partial(A.copy())

# Imprimir los resultados fuera de la función
print("Pivoteo Parcial - Solución:", x_partial)
print("Operaciones:", operations_partial)
print("Tiempo:", time_partial, "segundos\n")
```

####**Algoritmo de eliminación Gaussiana: Pivoteo Total**

```python
def gaussian_elimination_total(A):
    n = A.shape[0]  # Obtener el número de filas de la matriz A
    start_time = time.time()  # Iniciar la medición del tiempo
    operations = 0  # Inicializar el contador de operaciones

    # Fase de eliminación hacia adelante con pivoteo total
    for i in range(n):
        # Encontrar la posición del valor máximo absoluto en el submatriz A[i:n, i:n]
        max_row, max_col = np.unravel_index(np.argmax(np.abs(A[i:n, i:n])), (n-i, n-i))
        max_row += i  # Ajustar la fila del máximo a la posición original en A
        max_col += i  # Ajustar la columna del máximo a la posición original en A

        # Intercambiar filas y columnas para colocar el pivote máximo en la posición A[i, i]
        A[[i, max_row]] = A[[max_row, i]]  # Intercambiar filas
        A[:, [i, max_col]] = A[:, [max_col, i]]  # Intercambiar columnas

        A[i] = A[i] / A[i, i]  # Dividir la fila i por el elemento diagonal para hacer A[i, i] = 1
        operations += 2 * (n - i)  # Contar las operaciones (división y asignación)

        for j in range(i + 1, n):
            A[j] = A[j] - A[j, i] * A[i]  # Eliminar el elemento A[j, i] haciendo la fila j = fila j - A[j, i] * fila i
            operations += 2 * (n - i)  # Contar las operaciones (multiplicación, resta y asignación)

    # Fase de sustitución hacia atrás
    x = np.zeros(n)  # Inicializar el vector de soluciones x con ceros
    for i in range(n - 1, -1, -1):  # Recorrer las filas desde la última hasta la primera
        x[i] = A[i, -1] - np.sum(A[i, i + 1:n] * x[i + 1:n])  # Calcular x[i] restando el producto de los coeficientes y las soluciones ya calculadas
        operations += n - i  # Contar las operaciones (multiplicación y resta)

    end_time = time.time()  # Finalizar la medición del tiempo
    return x, operations, end_time - start_time  # Devolver el vector de soluciones, el número de operaciones y el tiempo de ejecución

# Llamar a la función y obtener resultados para el pivoteo total
x_total, operations_total, time_total = gaussian_elimination_total(A.copy())  # Usar una copia de A para no modificar la original

# Imprimir los resultados
print("Pivoteo Total - Solución:", x_total)  # Imprimir la solución del sistema de ecuaciones
print("Operaciones:", operations_total)  # Imprimir el número de operaciones realizadas
print("Tiempo:", time_total, "segundos\n")  # Imprimir el tiempo de ejecución
```

####**Algoritmo de eliminación: Pivoteo Escalonado**

```python
def gaussian_elimination_scaled(A):
    """
    Realiza la eliminación gaussiana con pivoteo escalonado en la matriz A.

    Parámetros:
    A (numpy array): Matriz aumentada del sistema de ecuaciones lineales.

    Retorna:
    tuple: Una tupla que contiene la solución del sistema, el número de operaciones realizadas y el tiempo de ejecución en segundos.
    """

    n = A.shape[0]  # Número de filas en la matriz A
    start_time = time.time()  # Guardar el tiempo de inicio
    operations = 0  # Inicializar contador de operaciones

    # Eliminación Gaussiana con pivoteo escalonado
    for i in range(n):
        # Calcular los factores de escala (máximo valor absoluto en cada fila desde i en adelante)
        scale = np.max(np.abs(A[i:n, :]), axis=1)
        # Encontrar la fila con el valor máximo en la columna i después de escalar
        max_row = np.argmax(np.abs(A[i:n, i]) / scale) + i
        # Intercambiar filas i y max_row
        A[[i, max_row]] = A[[max_row, i]]
        # Normalizar la fila i para hacer el pivote igual a 1
        A[i] = A[i] / A[i, i]
        operations += 2 * (n - i)  # Contar operaciones de la normalización

        # Eliminar elementos debajo del pivote en la columna i
        for j in range(i + 1, n):
            A[j] = A[j] - A[j, i] * A[i]
            operations += 2 * (n - i)  # Contar operaciones de la eliminación

    # Sustitución hacia atrás
    x = np.zeros(n)  # Inicializar vector de solución
    for i in range(n - 1, -1, -1):
        # Calcular el valor de x[i] usando los valores ya calculados de x[i+1] a x[n-1]
        x[i] = A[i, -1] - np.sum(A[i, i + 1:n] * x[i + 1:n])
        operations += n - i  # Contar operaciones de la sustitución

    end_time = time.time()  # Guardar el tiempo de finalización

    # Devolver la solución, el número de operaciones y el tiempo de ejecución
    return x, operations, end_time - start_time

# Ejemplo de uso con la misma matriz A
A = np.array([
    [0.04, 0.01, 0.01, 0.06],  # Primera ecuación: 0.04x1 + 0.01x2 + 0.01x3 = 0.06
    [0.2, 0.5, 0.2, 0.3],      # Segunda ecuación: 0.2x1 + 0.5x2 - 0.2x3 = 0.3
    [1, 2, 4, 11]              # Tercera ecuación: 1x1 + 2x2 + 4x3 = 11
], dtype=float)

# Llamar a la función y obtener resultados
x_scaled, operations_scaled, time_scaled = gaussian_elimination_scaled(A.copy())

# Imprimir los resultados fuera de la función
print("Pivoteo Escalonado - Solución:", x_scaled)
print("Operaciones:", operations_scaled)
print("Tiempo:", time_scaled, "segundos\n")
```

El pivoteo parcial puede ser más eficiente en términos de calidad de la solución y estabilidad numérica, especialmente en sistemas grandes o mal condicionados, a pesar de tener más operaciones. Para sistemas de ecuaciones más amplios, las diferencias en el número de operaciones y el tiempo de ejecución serán más pronunciadas.
El método sin pivoteo es más rápido porque no realiza estas operaciones adicionales, pero puede ser menos preciso y estable en sistemas grandes.

```python
# Definir las funciones de eliminación gaussiana

def gaussian_elimination(A):
    n = A.shape[0]
    operations = 0

    for i in range(n):
        A[i] = A[i] / A[i, i]
        operations += 2 * (n - i)
        for j in range(i + 1, n):
            A[j] = A[j] - A[j, i] * A[i]
            operations += 2 * (n - i)

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = A[i, -1] - np.sum(A[i, i + 1:n] * x[i + 1:n])
        operations += n - i

    return x, operations

def gaussian_elimination_partial(A):
    n = A.shape[0]
    operations = 0

    for i in range(n):
        max_row = np.argmax(np.abs(A[i:n, i])) + i
        A[[i, max_row]] = A[[max_row, i]]
        operations += 2 * n

        pivot = A[i, i]
        A[i] /= pivot
        operations += n

        for j in range(i + 1, n):
            factor = A[j, i]
            A[j] -= factor * A[i]
            operations += n

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = A[i, -1] - np.sum(A[i, i + 1:n] * x[i + 1:n])
        operations += n

    return x, operations

def gaussian_elimination_total(A):
    n = A.shape[0]
    operations = 0

    for i in range(n):
        max_row, max_col = np.unravel_index(np.argmax(np.abs(A[i:n, i:n])), (n-i, n-i))
        max_row += i
        max_col += i
        A[[i, max_row]] = A[[max_row, i]]
        A[:, [i, max_col]] = A[:, [max_col, i]]
        A[i] = A[i] / A[i, i]
        operations += 2 * (n - i)

        for j in range(i+1, n):
            A[j] = A[j] - A[j, i] * A[i]
            operations += 2 * (n - i)

    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = A[i, -1] - np.sum(A[i, i+1:n] * x[i+1:n])
        operations += n - i

    return x, operations

def gaussian_elimination_scaled(A):
    n = A.shape[0]
    operations = 0

    for i in range(n):
        scale = np.max(np.abs(A[i:n, :]), axis=1)
        max_row = np.argmax(np.abs(A[i:n, i]) / scale) + i
        A[[i, max_row]] = A[[max_row, i]]
        A[i] = A[i] / A[i, i]
        operations += 2 * (n - i)

        for j in range(i + 1, n):
            A[j] = A[j] - A[j, i] * A[i]
            operations += 2 * (n - i)

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = A[i, -1] - np.sum(A[i, i + 1:n] * x[i + 1:n])
        operations += n - i

    return x, operations

def medir_rendimiento(func, A):
    start_time = time.time()
    x, operations = func(A.copy())
    end_time = time.time()
    tiempo_ejecucion = end_time - start_time
    return x, operations, tiempo_ejecucion

# Ejemplo de uso con la misma matriz A
A = np.array([
    [0.04, 0.01, 0.01, 0.06],
    [0.2, 0.5, 0.2, 0.3],
    [1, 2, 4, 11]
], dtype=float)

# Medir rendimiento de los diferentes métodos
results = []

# Sin pivoteo
x, operations, exec_time = medir_rendimiento(gaussian_elimination, A.copy())
results.append(["Sin Pivoteo", operations, exec_time])

# Pivoteo parcial
x_partial, operations_partial, time_partial = medir_rendimiento(gaussian_elimination_partial, A.copy())
results.append(["Pivoteo Parcial", operations_partial, time_partial])

# Pivoteo total
x_total, operations_total, time_total = medir_rendimiento(gaussian_elimination_total, A.copy())
results.append(["Pivoteo Total", operations_total, time_total])

# Pivoteo escalonado
x_scaled, operations_scaled, time_scaled = medir_rendimiento(gaussian_elimination_scaled, A.copy())
results.append(["Pivoteo Escalonado", operations_scaled, time_scaled])

# Crear un DataFrame con los resultados
df = pd.DataFrame(results, columns=["Método", "Operaciones", "Tiempo (segundos)"])

# Imprimir la tabla de resultados
print(df)
```

# SISTEMAS DE ECUACIONES NO LINEALES


##TEORÍA DE RESOLUCIÓN DE SISTEMAS DE ECUACIONES NO LINEALES

Las ecuaciones no lineales son aquellas en las que la función $\large f( x)$ no es lineal. Por lo que para resolver estas ecuaciones implica encontrar las raíces donde $\large f( x) =\ 0$.

Este tipo de ecuaciones aparece frecuentemente en la modelización de fenómenos científicos y de ingeniería.

Dentro de la teoría de la solución de ecuaciones no lineales y presenta varios métodos numéricos.

\\
A continuación se puntualizará en cada uno:

###**Métodos numéricos para resolver ecuaciones no lineales**

Se componen de tres métodos principales:

1.   Método de la Bisección
2.   Método de Newton
3.   Método de la Secante


###**1.   Método de la Bisección**
El método de la bisección se basa en el Teorema del Valor Intermedio, que garantiza que si una función continua cambia de signo en un intervalo $\large [ a,b]$ , entonces tiene al menos una raíz en ese intervalo.

####**Algoritomo del método de la bisección**

*   Se toma un intervalo $\large [ a,b]$ donde $\large f( a)$ y $\large f( b)$ tienen signos contrarios

*   Se calcula el punto medio $\large c=\frac{a+b}{2}$

*   Se evalua $\large f( c)$

      - Si $\large f( c) =\ 0$, entonces $\large c$ es la raiz.

      - Si $\large f( a) \cdotp f( c) < 0$, la raíz está en $\large [ a,c]$; de lo contrario está en $\large [ c,b]$

*   Se repite el proceso hasta que el intervalo $\large [ a,b]$ sea suficientemente pequeño.

Este método es robusto y siempre converge si la función es continua en  $\large [ a,b]$, pero su convergencia es lineal y puede ser lenta con un comportamiento inestable

\\

###**2.   Método de Newton**
El método de Newton es un método iterativo que usa la derivada de la función para encontrar las raíces. A partir de una estimación inicial $\large x_{i}$, se aproxima la función mediante su serie de Taylor y se resuelve para encontrar una mejor aproximación.

\\
####**Fórmula del método de Newton**

$\large x_{i+1} \ =\ x_{i} -\frac{f( x_{i})}{f´( x_{i})}$

\\
Este método converge rápidamente si la estimación inicial está cerca de la raíz, pero no siempre converge y puede fallar si la derivada es cero o si la función no es bien comportada cerca de la raíz.

\\


###**3.   Método de la Secante**
El método de la secante es similar al método de Newton pero no requiere la derivada de la función. En su lugar, usa una aproximación de la derivada calculada con dos estimaciones anteriores.

\\
####**Fórmula del método de la secante**

$\large x_{i+1} \ =\ x_{i} -f( \ x_{i}) \cdotp \frac{\ x_{i} -x_{i-1}}{f( \ x_{i}) -f( x_{i-1})}$

\\
Para este método es necesario conocer los valores de $\large x_{i}$ y de $\large x_{i-1}$ para calcular $\large x_{i+1}$ así mismo, que $\large f( x_{i-1}) \ \neq \ f( \ x_{i})$

Los cuales representan:

*   $\large x_{i-1}$: La nueva aproximación a la raíz.

\\
*   $\large x_{i}$: La aproximación actual a la raíz en la iteración i

\\
*   $\large x_{i-1}$: La aproximación a la raíz en la iteración anterior i-1.

\\
*   $\large f( \ x_{i})$: El valor de la función en la aproximación actual  $\large x_{i}$

\\
*   $\large f( x_{i-1})$: El valor de la función en la aproximación anterior.

\\
*   $\large f( \ x_{i}) -f( x_{i-1})$: Diferencia en los valores de la función correspondientes a las dos últimas aproximaciones. Se debe tener cuidado al hacer la división para $\large f( \ x_{i}) -f( x_{i-1})$, puesto
que al acercarse a la raíz este valor será muy pequeño y podemos producir un
desbordamiento aritmético

\\
*   $\large \frac{\ x_{i} -x_{i-1}}{f( \ x_{i}) -f( x_{i-1})}$ Este cociente es una aproximación de la derivada de $\large f( x)$ en el intervalo $\large [ x_{i-1} ,\ x_{i}]$ En el método de la secante, se usa esta aproximación en lugar de la derivada exacta $\large \ f´( \ x)$

\\
Este método puede ser más práctico que el de Newton cuando la derivada no es fácilmente calculable, ya que no necesita conocerse su primera derivada. Sin embargo, al igual que el método de Newton, no siempre converge y su comportamiento puede ser errático.

\\


---

##**Conclusión**

Cada uno de estos métodos tiene sus propias ventajas y desventajas. El método de la bisección es confiable pero lento, el método de Newton es rápido pero puede no converger, y el método de la secante es una buena alternativa cuando la derivada no está disponible. Por lo que elegir el método adecuado depende de las características específicas del sistema de ecuaciones no lineales a resolver y de la función a resolver.

### RESOLUCIÓN NUMÉRICA DEL SISTEMA DE ECUACIÓN NO LINEAL UTILIZANDO EL METODO DE NEWTON-RAPSON
## 1. Planteamiento del Problema

Resolver el siguiente sistema de ecuaciones no lineales:

$$
\begin{cases}
\sin(x^2) - y = -2 \\
x^2 - y^2 = -5
\end{cases}
$$

## 2. Método de Newton-Raphson

El método de Newton-Raphson para sistemas de ecuaciones se define por la siguiente fórmula iterativa:

\\
$$\mathbf{x}_{k+1} = \mathbf{x}_k - J^{-1}(\mathbf{x}_k) \mathbf{f}(\mathbf{x}_k)$$

\\
Donde $\mathbf{x}_k = (x_k, y_k)$ es el vector de variables en la iteración $k$, $J$ es la matriz Jacobiana, y $\mathbf{f}$ es el vector de funciones.

\\
## 3. Preparación

Reescribimos el sistema en la forma $\mathbf{f}(\mathbf{x}) = \mathbf{0}$:

$$
\begin{aligned}
f_1(x,y) &= \sin(x^2) - y + 2 = 0 \\
f_2(x,y) &= x^2 - y^2 + 5 = 0
\end{aligned}
$$

\\
La matriz Jacobiana es:

$$
J = \begin{pmatrix}
\frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y} \\
\frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y}
\end{pmatrix} =
\begin{pmatrix}
2x \cos(x^2) & -1 \\
2x & -2y
\end{pmatrix}
$$

## 4. Iteraciones

### Iteración 0 (Punto inicial)

Elegimos como punto inicial:

$$\mathbf{x}_0 = \begin{pmatrix} 1.0 \\ 1.0 \end{pmatrix}$$

\\
### Iteración 1

Calculamos $\mathbf{f}(\mathbf{x}_0)$:

$$
\mathbf{f}(\mathbf{x}_0) = \begin{pmatrix} \sin(1^2) - 1 + 2 \\ 1^2 - 1^2 + 5 \end{pmatrix} = \begin{pmatrix} 1.8415 \\ 5 \end{pmatrix}
$$

\\
Calculamos $J(\mathbf{x}_0)$:

$$
J(\mathbf{x}_0) = \begin{pmatrix} 2(1)\cos(1^2) & -1 \\ 2(1) & -2(1) \end{pmatrix} = \begin{pmatrix} 1.0806 & -1 \\ 2 & -2 \end{pmatrix}
$$

\\
Calculamos $J^{-1}(\mathbf{x}_0)$:

$$
J^{-1}(\mathbf{x}_0) = \begin{pmatrix} -0.9254 & -0.5000 \\ -0.5000 & -0.5000 \end{pmatrix}
$$

\\
Aplicamos la fórmula de Newton-Raphson:

$$
\begin{aligned}
\mathbf{x}_1 &= \mathbf{x}_0 - J^{-1}(\mathbf{x}_0) \mathbf{f}(\mathbf{x}_0) \\
&= \begin{pmatrix} 1.0 \\ 1.0 \end{pmatrix} - \begin{pmatrix} -0.9254 & -0.5000 \\ -0.5000 & -0.5000 \end{pmatrix} \begin{pmatrix} 1.8415 \\ 5 \end{pmatrix} \\
&= \begin{pmatrix} 1.0 \\ 1.0 \end{pmatrix} - \begin{pmatrix} -3.2037 \\ -3.4207 \end{pmatrix} = \begin{pmatrix} 4.2037 \\ 4.4207 \end{pmatrix}
\end{aligned}
$$

\\
### Iteración 2

Calculamos $\mathbf{f}(\mathbf{x}_1)$:

$$
\mathbf{f}(\mathbf{x}_1) = \begin{pmatrix} \sin(4.2037^2) - 4.4207 + 2 \\ 4.2037^2 - 4.4207^2 + 5 \end{pmatrix} = \begin{pmatrix} -2.4177 \\ -1.9456 \end{pmatrix}
$$

\\
Calculamos $J(\mathbf{x}_1)$:

$$
J(\mathbf{x}_1) = \begin{pmatrix} 2(4.2037)\cos(4.2037^2) & -1 \\ 2(4.2037) & -2(4.4207) \end{pmatrix} = \begin{pmatrix} -8.0557 & -1 \\ 8.4074 & -8.8414 \end{pmatrix}

##IMPLEMENTACIÓN DEL MÉTODO NUMÉRICO Y RESOLUCIÓN DEL SISTEMA DE ECUACIONES NO LINEALES

```python

def f1(x, y):
    return np.sin(x**2) - y + 2

def f2(x, y):
    return x**2 - y**2 + 5

def F(X):
    x, y = X
    return np.array([f1(x, y), f2(x, y)])

def jacobian(X):
    x, y = X
    return np.array([[2*np.cos(x**2), -1],
                     [2*x, -2*y]])

def newton_raphson(F, jacobian, X0, tol=1e-6, max_iter=100):
    X = X0
    for i in range(max_iter):
        dX = np.linalg.solve(jacobian(X), -F(X))
        X = X + dX
        if np.linalg.norm(dX) < tol:
            return X, i
    return X, max_iter

# Punto inicial
X0 = np.array([1.0, 1.0])

# Aplicar Newton-Raphson
solution, iterations = newton_raphson(F, jacobian, X0)

print(f"Solución: x = {solution[0]:.4f}, y = {solution[1]:.4f}")
print(f"Iteraciones: {iterations}")
print(f"Norma del error: {np.linalg.norm(F(solution)):.2e}")
```

##CONCLUSIONES SOBRE LOS CRITERIOS PARA ENCONTRAR TODAS LAS RAICES DEL SISTEMA DE ECUACIONES NO LINEALES

En la resolución de sistemas de ecuaciones no lineales, como el que hemos abordado utilizando el método de Newton-Raphson, es fundamental considerar varios criterios y conclusiones importantes:

\\
1.  **Existencia de Raíces:** Es crucial determinar si el sistema tiene solución. Esto puede verificarse analizando las funciones originales y asegurándose de que hay intersecciones donde ambas ecuaciones se igualan a cero. En nuestro caso, las funciones $f1(x,y)$ y $f2(x,y)$ deben tener puntos donde se crucen.

\\
2.   **Condiciones Iniciales:** La elección de las condiciones iniciales $(x0,y0)$
es crucial. Pueden afectar la convergencia del método. Deben seleccionarse cuidadosamente cerca de la solución esperada para asegurar la convergencia rápida y evitar mínimos locales o puntos de inflexión que podrían conducir a divergencia.

\\
3.   **Convergencia:** El método de Newton-Raphson converge localmente cerca de la raíz cuando se cumplen ciertas condiciones (como la inversibilidad y continuidad del Jacobiano cerca de la raíz). En nuestro caso, hemos visto cómo las iteraciones convergen rápidamente a la solución esperada después de unas pocas iteraciones.

\\
4. **Iteraciones y Precisión:** La precisión de la solución depende del número de iteraciones y del criterio de convergencia establecido. En nuestro ejemplo, hemos realizado 7 iteraciones para alcanzar una solución precisa. Aumentar o disminuir el número de iteraciones puede afectar la precisión y eficiencia del método.  

\\
5.   **Verificación de la Solución:** Después de obtener una solución aproximada, siempre es crucial verificarla sustituyendo los valores encontrados en las funciones originales $f1(x,y)$ y $f2(x,y)$.
Los resultados deben estar lo más cerca posible de cero para confirmar que se ha encontrado una raíz válida.

\\
En resumen, al resolver sistemas de ecuaciones no lineales con el método de Newton-Raphson, se deben considerar estos criterios para garantizar la convergencia y precisión de las soluciones obtenidas. La elección adecuada de condiciones iniciales y la verificación cuidadosa son pasos fundamentales para encontrar todas las raíces del sistema de manera efectiva.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros
h = 0.01  # Tamaño de paso
t = np.arange(-0.25, 0.25, h)  # Intervalo de tiempo
n = len(t)

# Condiciones iniciales para cada método
y1_euler, y2_euler, y3_euler = [1], [0], [0]  # Método de Euler
y1_heun, y2_heun, y3_heun = [1], [0], [0]  # Método de Heun
y1_rk, y2_rk, y3_rk = [1], [0], [0]  # Método de Runge-Kutta de cuarto orden

# Solución de referencia (en este caso, se asume que es la solución de Runge-Kutta para alta precisión)
y1_exact = [1]
y2_exact = [0]
y3_exact = [0]

# Funciones de derivadas
def f1(y2):
    return y2

def f2(y3):
    return y3

def f3(y1, y2, y3):
    return -4 * y3 - 75 * y2 + 378 * y1

# Calculo la solución de referencia utilizando RK4 con un tamaño de paso más pequeño (h/10)
h_ref = h / 10
for i in range(int(n * 10) - 1):
    k1_y1 = h_ref * f1(y2_exact[-1])
    k1_y2 = h_ref * f2(y3_exact[-1])
    k1_y3 = h_ref * f3(y1_exact[-1], y2_exact[-1], y3_exact[-1])

    k2_y1 = h_ref * f1(y2_exact[-1] + k1_y2 / 2)
    k2_y2 = h_ref * f2(y3_exact[-1] + k1_y3 / 2)
    k2_y3 = h_ref * f3(y1_exact[-1] + k1_y1 / 2, y2_exact[-1] + k1_y2 / 2, y3_exact[-1] + k1_y3 / 2)

    k3_y1 = h_ref * f1(y2_exact[-1] + k2_y2 / 2)
    k3_y2 = h_ref * f2(y3_exact[-1] + k2_y3 / 2)
    k3_y3 = h_ref * f3(y1_exact[-1] + k2_y1 / 2, y2_exact[-1] + k2_y2 / 2, y3_exact[-1] + k2_y3 / 2)

    k4_y1 = h_ref * f1(y2_exact[-1] + k3_y2)
    k4_y2 = h_ref * f2(y3_exact[-1] + k3_y3)
    k4_y3 = h_ref * f3(y1_exact[-1] + k3_y1, y2_exact[-1] + k3_y2, y3_exact[-1] + k3_y3)

    y1_exact.append(y1_exact[-1] + (1 / 6) * (k1_y1 + 2 * k2_y1 + 2 * k3_y1 + k4_y1))
    y2_exact.append(y2_exact[-1] + (1 / 6) * (k1_y2 + 2 * k2_y2 + 2 * k3_y2 + k4_y2))
    y3_exact.append(y3_exact[-1] + (1 / 6) * (k1_y3 + 2 * k2_y3 + 2 * k3_y3 + k4_y3))

y1_exact = y1_exact[::10]
y2_exact = y2_exact[::10]
y3_exact = y3_exact[::10]

# Método de Euler
for i in range(n - 1):
    y1_euler.append(y1_euler[-1] + h * f1(y2_euler[-1]))
    y2_euler.append(y2_euler[-1] + h * f2(y3_euler[-1]))
    y3_euler.append(y3_euler[-1] + h * f3(y1_euler[-1], y2_euler[-1], y3_euler[-1]))

# Método de Heun
for i in range(n - 1):
    y1_predict = y1_heun[-1] + h * f1(y2_heun[-1])
    y2_predict = y2_heun[-1] + h * f2(y3_heun[-1])
    y3_predict = y3_heun[-1] + h * f3(y1_heun[-1], y2_heun[-1], y3_heun[-1])

    y1_heun.append(y1_heun[-1] + (h / 2) * (f1(y2_heun[-1]) + f1(y2_predict)))
    y2_heun.append(y2_heun[-1] + (h / 2) * (f2(y3_heun[-1]) + f2(y3_predict)))
    y3_heun.append(y3_heun[-1] + (h / 2) * (f3(y1_heun[-1], y2_heun[-1], y3_heun[-1]) + f3(y1_predict, y2_predict, y3_predict)))

# Método de Runge-Kutta de cuarto orden (RK4)
for i in range(n - 1):
    k1_y1 = h * f1(y2_rk[-1])
    k1_y2 = h * f2(y3_rk[-1])
    k1_y3 = h * f3(y1_rk[-1], y2_rk[-1], y3_rk[-1])

    k2_y1 = h * f1(y2_rk[-1] + k1_y2 / 2)
    k2_y2 = h * f2(y3_rk[-1] + k1_y3 / 2)
    k2_y3 = h * f3(y1_rk[-1] + k1_y1 / 2, y2_rk[-1] + k1_y2 / 2, y3_rk[-1] + k1_y3 / 2)

    k3_y1 = h * f1(y2_rk[-1] + k2_y2 / 2)
    k3_y2 = h * f2(y3_rk[-1] + k2_y3 / 2)
    k3_y3 = h * f3(y1_rk[-1] + k2_y1 / 2, y2_rk[-1] + k2_y2 / 2, y3_rk[-1] + k2_y3 / 2)

    k4_y1 = h * f1(y2_rk[-1] + k3_y2)
    k4_y2 = h * f2(y3_rk[-1] + k3_y3)
    k4_y3 = h * f3(y1_rk[-1] + k3_y1, y2_rk[-1] + k3_y2, y3_rk[-1] + k3_y3)

    y1_rk.append(y1_rk[-1] + (1 / 6) * (k1_y1 + 2 * k2_y1 + 2 * k3_y1 + k4_y1))
    y2_rk.append(y2_rk[-1] + (1 / 6) * (k1_y2 + 2 * k2_y2 + 2 * k3_y2 + k4_y2))
    y3_rk.append(y3_rk[-1] + (1 / 6) * (k1_y3 + 2 * k2_y3 + 2 * k3_y3 + k4_y3))

# Calcular errores absolutos
error_euler = np.abs(np.array(y1_euler) - np.array(y1_exact))
error_heun = np.abs(np.array(y1_heun) - np.array(y1_exact))
error_rk = np.abs(np.array(y1_rk) - np.array(y1_exact))

# Graficar errores
plt.figure(figsize=(10, 6))
plt.plot(t, error_euler, label='Error Euler', linestyle='--', marker='o')
plt.plot(t, error_heun, label='Error Heun', linestyle='--', marker='o')
plt.plot(t, error_rk, label='Error Runge-Kutta (RK4)', linestyle='--', marker='o')
plt.title('Comparación de Errores de Métodos Numéricos', fontsize=14)
plt.xlabel('Tiempo $t$', fontsize=12)
plt.ylabel('Error Absoluto', fontsize=12)
plt.yscale('log')  # Escala logarítmica para visualizar mejor los errores pequeños
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos la función de la ecuación diferencial
def dydx(x, y):
    return (2 * x * y) / (3 * x**2 - 6 * y**2)

# Condiciones iniciales y parámetros
x0, y0 = 1, 2
h = 0.1
n_steps = 3  # Número de pasos (cambiar según sea necesario)
x_values = [x0 + i * h for i in range(n_steps + 1)]

# Solución analítica aproximada para referencia
sol = solve_ivp(dydx, [x0, x0 + n_steps * h], [y0], method='RK45', t_eval=x_values)
y_exact = sol.y[0]

# Método de Euler
y_euler = [y0]
for i in range(n_steps):
    y_next = y_euler[-1] + h * dydx(x_values[i], y_euler[-1])
    y_euler.append(y_next)

# Método de Heun
y_heun = [y0]
for i in range(n_steps):
    fxy = dydx(x_values[i], y_heun[-1])
    y_predict = y_heun[-1] + h * fxy
    y_correct = y_heun[-1] + (h / 2) * (fxy + dydx(x_values[i + 1], y_predict))
    y_heun.append(y_correct)

# Método de Runge-Kutta de segundo orden (RK2)
y_rk2 = [y0]
for i in range(n_steps):
    k1 = h * dydx(x_values[i], y_rk2[-1])
    k2 = h * dydx(x_values[i] + h / 2, y_rk2[-1] + k1 / 2)
    y_rk2.append(y_rk2[-1] + k2)

# Cálculo de errores absolutos
error_euler = np.abs(np.array(y_euler) - np.array(y_exact))
error_heun = np.abs(np.array(y_heun) - np.array(y_exact))
error_rk2 = np.abs(np.array(y_rk2) - np.array(y_exact))

# Graficar errores
plt.figure(figsize=(10, 6))
plt.plot(x_values, error_euler, label='Error Euler', marker='o')
plt.plot(x_values, error_heun, label='Error Heun', marker='o')
plt.plot(x_values, error_rk2, label='Error Runge-Kutta (RK2)', marker='o')
plt.title('Análisis de Errores de Métodos Numéricos')
plt.xlabel('x')
plt.ylabel('Error Absoluto')
plt.yscale('log')  # Escala logarítmica para mejor visualización
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend()
plt.show()
```
