```python
import numpy as np
import matplotlib.pyplot as plt

# Habilitar el renderizado con LaTeX
plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "serif"  # Usa una fuente serif (similar a Times New Roman)

# Definimos la función
t = np.linspace(-4, 1, 50)
f_t = 5 * np.exp(-t) * np.cos(10 * t)  # Función f(t)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(t, f_t, label=r"$f(t) = 5e^{-t}\:\cos10t \; \in[-4,1]$", color="b")

# Agregar etiquetas y título
plt.title(r" $f(t) = 5e^{-t}\:\cos10t \; \in[-4,1]$, 50 puntos", fontsize=14, loc="left")
plt.title("Richard Pante", fontsize=14, loc="right")
plt.xlabel("x = t", fontsize=12)
plt.ylabel("y = f(t)", fontsize=12)

# Añadir grillas principales y secundarias
plt.grid(which='major', color='gray', linestyle='-', linewidth=0.5)
plt.grid(which='minor', color='lightgray', linestyle='--', linewidth=0.3)
plt.minorticks_on()  # Activar marcas secundarias

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

$\sqrt{7}≈3\sum_{n=0}

```python
import math
def binomial_coeff(x, n):
    if n == 0:
        return 1  # Caso base
    coeff = 1
    for i in range(n):
        coeff *= (x - i) / (i + 1)
    return coeff

def raiz7Aprox(N):
    raiz7 = 0
    for n in range(N + 1):
        raiz7 += binomial_coeff(0.5, n) * ((-1 / 4) ** n)
    return 3 * raiz7

def N_raiz7(precision):
    real_value = math.sqrt(7)
    N = 0
    while True:
        approx_value = raiz7Aprox(N)
        error = abs(approx_value - real_value)
        if error < precision:
            break
        N += 1
    return N

print("funcion 1")
print(raiz7Aprox(0))
print(raiz7Aprox(1))
print(raiz7Aprox(5))
print(raiz7Aprox(10))
print(raiz7Aprox(15))
print(raiz7Aprox(25))

print("funcion 2")
print(N_raiz7(0.1))
print(N_raiz7(0.01))
print(N_raiz7(0.001))
print(N_raiz7(0.000001))
print(N_raiz7(0.00000001))
```
