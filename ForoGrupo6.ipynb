##3. Serie de Taylor

##4. Generar las series para  n = 1, 3, 5, 7, 9, 11

Se ha obtenido la función aplicada a $\large n = 1, 3, 5, 7, 9, 11$; en base a la formula inicial de Taylor

$\large f( x) \simeq \frac{f^{( 1)}( x_{0})}{1!}( x-x_{0}) +\frac{f^{( 3)}( x_{0})}{3!}( x-x_{0})^{3} +\frac{f^{( 5)}( x_{0})}{5!}( x-x_{0})^{5} +\frac{f^{( 7)}( x_{0})}{7!}( x-x_{0})^{7} +\frac{f^{( 9)}( x_{0})}{9!}( x-x_{0})^{9} +\frac{f^{( 11)}( x_{0})}{11!}( x-x_{0})^{11}$

Se procede a reemplazar en la función principal aplicando a cada caso n

      Para n = 1
$\large f( x) =\sum _{n=1}^{\infty }\frac{f^{( 1)}( c)}{1!}( x-x_{0})^{n} =\frac{1}{1!}( x-0)^{1} =x\ $

      Para n = 3
$\large f( x) =\sum _{n=3}^{\infty }\frac{f^{( 1)}( c)}{1!}( x-x_{0})^{n} +\ \frac{f^{( 3)}( c)}{3!}( x-x_{0})^{3} =x+\frac{-6}{3!}( x-0)^{3} =x+\frac{-6}{3!} x^{3} \ $

      Para n = 5
$\large f( x) =\sum _{n=5}^{\infty }\frac{f^{( 1)}( c)}{1!}( x-x_{0})^{n} +\ \frac{f^{( 3)}( c)}{3!}( x-x_{0})^{3} +\frac{f^{( 5)}( c)}{5!}( x-x_{0})^{5} =x+\frac{-6}{3!} x^{3} +\frac{120}{5!}( x-0)^{5} =x+\frac{-6}{3!} x^{3} +\frac{120}{5!} x^{5}$

```python

def taylor_series(f, x0, n):

    x = sp.Symbol('x')  # Declara 'x' como una variable simbólica.
    series = 0  # Inicializa la serie de Taylor en 0.
    for i in range(n+1):  # Itera 'n' veces.
        deriv = sp.diff(f, x, i)  # Calcula la i-ésima derivada de 'f' respecto a 'x'.
        term = deriv.subs(x, x0) / sp.factorial(i)  # Evalúa la derivada en x0 y divide por i!.
        series += term * (x - x0)**i  # Añade el término actual a la serie.
    return series  # Devuelve la serie de Taylor completa.

# Define la función f(x)
x = sp.Symbol('x')  # Declara 'x' como variable simbólica.
f = x / (1 + x**2)  # Define la función f(x).

# Punto central para la expansión
c = 0  # Establece c como el punto central de la serie.




# Cantidad de términos para las series
terms = [1, 3, 5, 7, 9, 11]  # Lista con diferentes números de términos para las series.

# Calcula y muestra las series de Taylor
for n in terms:  # Itera sobre cada valor en 'terms'.
    series = taylor_series(f, c, n)  # Calcula la serie de Taylor con n términos.
    print(f"Serie de Maclaurin para n = {n}:")  # Imprime el número de términos.
    print(series)  # Imprime la serie de Taylor.
    print()  # Imprime una línea en blanco para separar los resultados.

'''
for a in range(20):
  #if, elif, for, while,
  print(a+1)
'''
```

Para exponer y comentar el gráfico unificado con las distintas aproximaciones de Taylor, es fundamental entender cómo cada aproximación se acerca progresivamente a la función original a medida que se incluyen más términos de la serie de Taylor. La función original parece estar representada únicamente por

Las funciones consideradas son:
* $f(x)=x$

* $f(x) = x - \frac{6}{3!}x^3 + \frac{120}{5!}x^5$

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt


# Definir las aproximaciones de Taylor
def taylor_approx_1(x):
    return x

def taylor_approx_2(x):
    return x - (6 / mt.factorial(3)) * x**3

def taylor_approx_3(x):
    return x - (6 / mt.factorial(3)) * x**3 + (120 / mt.factorial(5)) * x**5

def taylor_approx_4(x):
    return x - (6 / mt.factorial(3)) * x**3 + (120 / mt.factorial(5)) * x**5 - (5040 / mt.factorial(7)) * x**7

def taylor_approx_5(x):
    return x - (6 / mt.factorial(3)) * x**3 + (120 / mt.factorial(5)) * x**5 - (5040 / mt.factorial(7)) * x**7 + (362880 / mt.factorial(9)) * x**9

def taylor_approx_6(x):
    return x - (6 / mt.factorial(3)) * x**3 + (120 / mt.factorial(5)) * x**5 - (5040 / mt.factorial(7)) * x**7 + (362880 / mt.factorial(9)) * x**9 - (39916800 / mt.factorial(11)) * x**11

# Crear los valores de x
x_values = np.linspace(-10 , 10, 400)

# Calcular los valores y para cada aproximación de Taylor
y_values_1 = taylor_approx_1(x_values)
y_values_2 = taylor_approx_2(x_values)
y_values_3 = taylor_approx_3(x_values)
y_values_4 = taylor_approx_4(x_values)
y_values_5 = taylor_approx_5(x_values)
y_values_6 = taylor_approx_6(x_values)

# Crear la figura
plt.figure(figsize=(10, 6))

plt.xlim(-10, 10)  # Limitar el eje X para evitar la divergencia
plt.ylim(-10, 10)  # Limitar el eje Y para evitar la divergencia

# Graficar cada aproximación de Taylor
plt.plot(x_values, y_values_1, label='$f(x) = x$', linewidth=2)
plt.plot(x_values, y_values_2, label='$f(x) = x - \\frac{6}{3!}x^3$', linewidth=2)
plt.plot(x_values, y_values_3, label='$f(x) = x - \\frac{6}{3!}x^3 + \\frac{120}{5!}x^5$', linewidth=2)
plt.plot(x_values, y_values_4, label='$f(x) = x - \\frac{6}{3!}x^3 + \\frac{120}{5!}x^5 - \\frac{5040}{7!}x^7$', linewidth=2)
plt.plot(x_values, y_values_5, label='$f(x) = x - \\frac{6}{3!}x^3 + \\frac{120}{5!}x^5 - \\frac{5040}{7!}x^7 + \\frac{362880}{9!}x^9$', linewidth=2)
plt.plot(x_values, y_values_6, label='$f(x) = x - \\frac{6}{3!}x^3 + \\frac{120}{5!}x^5 - \\frac{5040}{7!}x^7 + \\frac{362880}{9!}x^9 - \\frac{39916800}{11!}x^{11}$', linewidth=2)

# Configurar el gráfico
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Aproximaciones de Taylor', loc="left")
plt.title('GRUPO 6', loc="right")
plt.legend()
plt.grid(which='major', axis='both', linestyle='-', linewidth=0.5)
plt.grid(which='minor', axis='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.axhline(0, color="black")
plt.axvline(0, color="black")

# Mostrar el gráfico
plt.show()
```
