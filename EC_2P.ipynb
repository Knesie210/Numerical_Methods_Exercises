## RESOLUCIÓN ANALÍTICA
$\displaystyle  \begin{array}{{>{\displaystyle}l}}

\left( 3x^{2} -6y^{2}\right)\frac{dy}{dx} \ =\ 2xy\\

\frac{dy}{dx} \ =\frac{2xy}{3\left( x^{2} -2y^{2}\right)}\\

\\
Se va a realizar un cambio de variable para simplificar la ecuación

\\
$\displaystyle v\ =\ \frac{y}{x}$

$\displaystyle y\ =vx$

$\displaystyle \frac{dy}{dx} =v+x\frac{dv}{dx}$

\\
Sustituyendo en la ecuacion original, se obtiene:

\\
$\displaystyle  \begin{array}{{>{\displaystyle}l}}
\left( 3x^{2} -6( vx)^{2}\right)\left( v+x\frac{dv}{dx}\right) =2x( vx)\\
\\
\left( 3x^{2} -6v^{2} x^{2}\right)\left( v+x\frac{dv}{dx}\right) =2vx^{2}
\end{array}$

\\
Factorizando $\displaystyle x^{2}$ del lado izquierdo se tiene:

\\
$\displaystyle x^{2}\left( 3-6v^{2}\right)\left( v+x\frac{dv}{dx}\right) =2vx^{2}$

\\
Se elimina  $\displaystyle x^{2}$ de ambos lados:

\\
$\displaystyle \left( 3-6v^{2}\right)\left( v+x\frac{dv}{dx}\right) =2v$

\\
Se procede a expandir y simplificar la ecuacion:

\\
$\displaystyle  \begin{array}{{>{\displaystyle}l}}
\left( 3v-6v^{3}\right) +x\left( 3-6v^{2}\right)\frac{dv}{dx} =2v\\
\\
x\left( 3-6v^{2}\right)\frac{dv}{dx} =2v-3v+6v^{3}\\

x\left( 3-6v^{2}\right)\frac{dv}{dx} =-v+6v^{3}
\end{array}$

\\
Se procede a resolver por metodos de integracion

\\
$\displaystyle \frac{dv}{dx} =\frac{-v+6v^{3}}{x\left( 3-6v^{2}\right)}$

\\
Se reorganiza la ecuacion para separar las variables $\displaystyle v$ y $\displaystyle x$

\\
$\displaystyle \frac{3-6v^{2}}{-v+6v^{3}} dv\ =\frac{1}{x} dx$

\\
Integracion del lado derecho en x

\\
$\displaystyle \int \frac{1}{x} dx=ln\mid x\mid +C$

\\
$\displaystyle  \begin{array}{{>{\displaystyle}l}}
\int \frac{3-6v^{2}}{-v+6v^{3}} dv\\
\\
\frac{3-6v^{2}}{-v\left( 1-2v^{2}\right)} dv
\end{array}$$ $

\\
Se descompondra la fracción de la siguiente manera:

\\
$\displaystyle \frac{3-6v^{2}}{-v\left( 1-2v^{2}\right)} =\frac{A}{v} +\frac{B}{1-2v^{2}}$

$\displaystyle 3-6v^{2} =A\left( 1-2v^{2}\right) -Bv$

Se igualan los coeficientes de los terminos similares para determinar A y B:

Para el termino:  $\displaystyle v:0\ =\ -B$, lo que implica que $\displaystyle B=0$

\\
Entonces la integral se reduce a:

\\
$\displaystyle \int \frac{3}{v} dv=3ln\mid v\mid $

La solucion general es:

\\
$\displaystyle 3ln\mid v\mid \ =ln\mid x\mid +C$

##IMPLEMENTACIÓN DE LOS MÉTODOS NUMÉRICOS

### MÉTODO EULER

```python
# Definimos la función f(x, y), que es la derivada dy/dx de la ecuación diferencial
def f(x, y):
    return (2 * x * y) / (3 * x**2 - 6 * y**2)

# Parámetros iniciales
x0 = 1  # Valor inicial de x
y0 = 2  # Valor inicial de y
h = 0.1  # Tamaño del paso
x_final = 2  # Valor final de x para detener la simulación
n = int((x_final - x0) / h)  # Número de iteraciones basado en el rango de x y el tamaño de paso

# Listas para almacenar valores de x e y durante las iteraciones
x_values_euler = [x0]  # Inicializa la lista de valores de x con el valor inicial
y_values_euler = [y0]  # Inicializa la lista de valores de y con el valor inicial

# Método de Euler para resolver la ecuación diferencial
x, y = x0, y0  # Inicialización de los valores de x e y
for i in range(n):  # Itera desde 0 hasta n-1 (n iteraciones)
    y = y + h * f(x, y)  # Calcula el nuevo valor de y usando el método de Euler
    x = x + h  # Incrementa x en el tamaño de paso h
    x_values_euler.append(x)  # Guarda el nuevo valor de x en la lista
    y_values_euler.append(y)  # Guarda el nuevo valor de y en la lista

# Graficar los resultados obtenidos con el método de Euler
plt.plot(x_values_euler, y_values_euler, label="Euler")  # Crea la gráfica de x vs y
plt.xlabel("x")  # Etiqueta del eje x
plt.ylabel("y")  # Etiqueta del eje y
plt.title("Solución con Método de Euler")  # Título del gráfico
plt.legend()  # Muestra la leyenda en la gráfica
plt.grid(True)  # Muestra una cuadrícula en la gráfica para mayor claridad
plt.show()  # Muestra la gráfica

# Imprimir resultados numéricos para cada par (x, y)
print("Resultados con el Método de Euler:")
for i in range(len(x_values_euler)):  # Itera sobre todos los valores calculados
    print(f"x = {x_values_euler[i]:.2f}, y = {y_values_euler[i]:.4f}")  # Imprime los valores de x e y
```

#### **Gráficos donde se comparan las 4 soluciones**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la ecuación diferencial con manejo de denominador cercano a cero
def dydx(y, x):
    denom = 3*x**2 - 6*y**2
    # Añadir un umbral mínimo para evitar divisiones por valores extremadamente pequeños
    if abs(denom) < 1e-6:  # Evitar denominador cercano a 0
        denom = 1e-6 if denom > 0 else -1e-6
    return (2*x*y)/denom

# Condición inicial
y0 = 2

# Intervalo de valores de x
x = np.arange(2, 20, 0.1)

# Método de Euler
def euler(dydx, y0, x):
    y = np.zeros_like(x)
    y[0] = y0
    for i in range(1, len(x)):
        h = x[i] - x[i-1]
        y[i] = y[i-1] + h * dydx(y[i-1], x[i-1])
    return y

# Método de Heun
def heun(dydx, y0, x):
    y = np.zeros_like(x)
    y[0] = y0
    for i in range(1, len(x)):
        h = x[i] - x[i-1]
        y_predict = y[i-1] + h * dydx(y[i-1], x[i-1])
        y[i] = y[i-1] + (h/2) * (dydx(y[i-1], x[i-1]) + dydx(y_predict, x[i]))
    return y

# Método de Runge-Kutta (RK2)
def rk2(dydx, y0, x):
    y = np.zeros_like(x)
    y[0] = y0
    for i in range(1, len(x)):
        h = x[i] - x[i-1]
        k1 = h * dydx(y[i-1], x[i-1])
        k2 = h * dydx(y[i-1] + 0.5 * k1, x[i-1] + 0.5 * h)
        y[i] = y[i-1] + k2
    return y

# Soluciones usando los métodos numéricos
y_euler = euler(dydx, y0, x)
y_heun = heun(dydx, y0, x)
y_rk2 = rk2(dydx, y0, x)

# Gráfico de las soluciones
plt.figure(figsize=(12, 8))

# Graficar las soluciones
plt.plot(x, y_euler, label='Método de Euler', linestyle='dotted')
plt.plot(x, y_heun, label='Método de Heun', linestyle='dashdot')
plt.plot(x, y_rk2, label='Método de Runge-Kutta (RK2)', linestyle='dashed')

plt.title(r'Comparación de Soluciones: Euler, Heun, y Runge-Kutta (RK2)' '\n(3x² - 6y²) dy/dx - 2xy = 0', pad=20)

# Etiquetas de ejes
plt.xlabel('x')
plt.ylabel('y(x)')

# Añadir el texto "GRUPO 6" en la esquina superior derecha
plt.text(19, max(y_euler.max(), y_heun.max(), y_rk2.max()) - 2, 'GRUPO 6', fontsize=12, ha='right')

# Leyenda
plt.legend()

# Añadir grillas mayores y menores
plt.grid(which='major', axis='both', linestyle='-', linewidth=0.7)
plt.grid(which='minor', axis='both', linestyle='--', linewidth=0.5)

# Ajuste del límite de y
plt.ylim(-10, max(y_euler.max(), y_heun.max(), y_rk2.max()) + 1)

# Mostrar la gráfica
plt.show()
```

# Resolución de la Ecuación Diferencial

## Sistema de Ecuaciones Diferenciales

$$
\begin{cases}
\frac{dy}{dx} = z \\[2ex]
\frac{dz}{dx} = \frac{2xy}{3x^2 - 6y^2}

$$

## 1. Cálculo de las Condiciones Iniciales

Dado: $y(1) = 2$

Para $z(1)$:

$$
\begin{align*}
(3\cdot1^2 - 6\cdot2^2) \cdot z(1) - 2\cdot1\cdot2 &= 0 \\
(3 - 24)z(1) - 4 &= 0 \\

## 2. Sistema Completo con Condiciones Iniciales

$$
\begin{cases}
\frac{dy}{dx} = z \\[2ex]

## 3. Desarrollo de $\frac{dz}{dx}$

### 3.1. Expresión inicial para $z$:

$$
z = \frac{2xy}{3x^2 - 6y^2}
$$

### 3.2. Derivamos $z$ respecto a $x$:

$$
\frac{dz}{dx} = \frac{d}{dx}\left[\frac{2xy}{3x^2 - 6y^2}\right]
$$

### 3.3. Aplicamos la regla del cociente:

$$
\frac{d}{dx}\left[\frac{u}{v}\right] = \frac{v\frac{du}{dx} - u\frac{dv}{dx}}{v^2}
$$

Donde $u = 2xy$ y $v = 3x^2 - 6y^2$

### 3.4. Resultado de aplicar la regla:

$$
\frac{dz}{dx} = \frac{\frac{d}{dx}(2xy) \cdot (3x^2 - 6y^2) - 2xy \cdot \frac{d}{dx}(3x^2 - 6y^2)}{(3x^2 - 6y^2)^2}
$$

### 3.5. Calculamos $\frac{d}{dx}(2xy)$ y $\frac{d}{dx}(3x^2 - 6y^2)$:

$$
\frac{d}{dx}(2xy) = 2x \cdot \frac{dy}{dx} + 2y = 2xz + 2y
$$

$$
\frac{d}{dx}(3x^2 - 6y^2) = 6x - 12y \cdot \frac{dy}{dx} = 6x - 12yz
$$

### 3.6. Sustituimos en la ecuación:

$$
\frac{dz}{dx} = \frac{(2xz + 2y)(3x^2 - 6y^2) - 2xy(6x - 12yz)}{(3x^2 - 6y^2)^2}
$$

### 3.7. Expandimos el numerador:

$$
\begin{align*}
(2xz + 2y)(3x^2 - 6y^2) &= 6x^3z - 12xy^2z + 6x^2y - 12y^3 \\
-2xy(6x - 12yz) &= -12x^2y + 24xy^2z
\end{align*}
$$

### 3.8. Sumamos los términos:

$$
\frac{dz}{dx} = \frac{6x^3z - 12xy^2z + 6x^2y - 12y^3 - 12x^2y + 24xy^2z}{(3x^2 - 6y^2)^2}
$$

### 3.9. Simplificamos:

$$
\frac{dz}{dx} = \frac{6x^3z + 12xy^2z + 6x^2y - 12y^3 - 12x^2y}{(3x^2 - 6y^2)^2}
$$

## 4. Verificación de la Solución Propuesta

Propuesta de solución:

$$
y = \frac{2}{x} \implies \frac{dy}{dx} = -\frac{2}{x^2}
$$

Ecuación original:

$$
(3x^2 - 6y^2) \cdot \frac{dy}{dx} - 2xy = 0
$$

Sustitución de $y$ y $\frac{dy}{dx}$ en la ecuación:

$$
\left(3x^2 - 6\left(\frac{2}{x}\right)^2\right) \cdot \left(-\frac{2}{x^2}\right) - 2x\left(\frac{2}{x}\right) = 0
$$

Simplificación del término $y^2$:

$$
\left(3x^2 - \frac{24}{x^2}\right) \cdot \left(-\frac{2}{x^2}\right) - \frac{4}{x} = 0
$$

Multiplicación de términos:

$$
\frac{-6x^4 + 48}{x^4} - \frac{4}{x} = 0
$$

Simplificación del denominador:

$$
\frac{-6x^3 + 48/x}{x^3} - \frac{4}{x} = 0
$$

Separación de términos:

$$
-6 + \frac{48}{x^4} - \frac{4}{x} = 0
$$

Esta igualdad se cumple para todo $x \neq 0$, verificando que $y = \frac{2}{x}$ es una solución de la ecuación diferencial original.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
h = 0.01  # Tamaño de paso (intervalo entre puntos)
t = np.arange(-0.25, 0.25, h)  # Intervalo de tiempo desde -0.25 hasta 0.25

# Condiciones iniciales para cada método
y1_euler = [1]  )# Valor inicial para y1 (z)
y2_euler = [0]  # Valor inicial para y2 (z')
y3_euler = [0]  # Valor inicial para y3 (z''

y1_heun = [1]
y2_heun = [0]
y3_heun = [0]

y1_rk = [1]
y2_rk = [0]
y3_rk = [0]

# Ecuación diferencial de tercer orden:
# z'''(t) + 4z''(t) + 75z'(t) - 378z(t) = 0
# Se convierte en un sistema de ecuaciones de primer orden:
# y1 = z(t)
# y2 = z'(t)
# y3 = z''(t)
# Entonces, el sistema es:
# y1' = y2
# y2' = y3
# y3' = -4*y3 - 75*y2 + 378*y1

# Método de Euler
for i in range(len(t) - 1):
    # Actualización de y1, y2, y3 usando el método de Euler
    y1_euler.append(y1_euler[-1] + h * y2_euler[-1])
    y2_euler.append(y2_euler[-1] + h * y3_euler[-1])
    y3_euler.append(y3_euler[-1] + h * (4 * y3_euler[-1] + 75 * y2_euler[-1] - 378 * y1_euler[-1]))

# Método de Heun
for i in range(len(t) - 1):
    # Paso predictor (Euler)
    y1_predict = y1_heun[-1] + h * y2_heun[-1]
    y2_predict = y2_heun[-1] + h * y3_heun[-1]
    y3_predict = y3_heun[-1] + h * (4 * y3_heun[-1] + 75 * y2_heun[-1] - 378 * y1_heun[-1])

    # Paso corrector (promedio de las pendientes del predictor y el valor actual)
    y1_heun.append(y1_heun[-1] + (h / 2) * (y2_heun[-1] + y2_predict))
    y2_heun.append(y2_heun[-1] + (h / 2) * (y3_heun[-1] + y3_predict))
    y3_heun.append(y3_heun[-1] + (h / 2) * ((4 * y3_heun[-1] + 75 * y2_heun[-1] - 378 * y1_heun[-1]) +
                                            (4 * y3_predict + 75 * y2_predict - 378 * y1_predict)))

# Método de Runge-Kutta de cuarto orden (RK4)
for i in range(len(t) - 1):
    # k1: Pendiente en el punto actual
    k1_y1 = h * y2_rk[-1]
    k1_y2 = h * y3_rk[-1]
    k1_y3 = h * (4 * y3_rk[-1] + 75 * y2_rk[-1] - 378 * y1_rk[-1])

    # k2: Pendiente en el punto medio (usando k1)
    k2_y1 = h * (y2_rk[-1] + k1_y2 / 2)
    k2_y2 = h * (y3_rk[-1] + k1_y3 / 2)
    k2_y3 = h * (4 * (y3_rk[-1] + k1_y3 / 2) + 75 * (y2_rk[-1] + k1_y2 / 2) - 378 * (y1_rk[-1] + k1_y1 / 2))

    # k3: Pendiente en el punto medio (usando k2)
    k3_y1 = h * (y2_rk[-1] + k2_y2 / 2)
    k3_y2 = h * (y3_rk[-1] + k2_y3 / 2)
    k3_y3 = h * (4 * (y3_rk[-1] + k2_y3 / 2) + 75 * (y2_rk[-1] + k2_y2 / 2) - 378 * (y1_rk[-1] + k2_y1 / 2))

    # k4: Pendiente en el punto final (usando k3)
    k4_y1 = h * (y2_rk[-1] + k3_y2)
    k4_y2 = h * (y3_rk[-1] + k3_y3)
    k4_y3 = h * (4 * (y3_rk[-1] + k3_y3) + 75 * (y2_rk[-1] + k3_y2) - 378 * (y1_rk[-1] + k3_y1))

    # Actualización usando la combinación de los valores de k1, k2, k3, y k4
    y1_rk.append(y1_rk[-1] + (1 / 6) * (k1_y1 + 2 * k2_y1 + 2 * k3_y1 + k4_y1))
    y2_rk.append(y2_rk[-1] + (1 / 6) * (k1_y2 + 2 * k2_y2 + 2 * k3_y2 + k4_y2))
    y3_rk.append(y3_rk[-1] + (1 / 6) * (k1_y3 + 2 * k2_y3 + 2 * k3_y3 + k4_y3))

# Imprimir resultados finales de los métodos
print("Resultados finales:")
print(f"Método de Euler: z({t[-1]}) = {y1_euler[-1]}")
print(f"Método de Heun: z({t[-1]}) = {y1_heun[-1]}")
print(f"Método de Runge-Kutta: z({t[-1]}) = {y1_rk[-1]}")

# Gráficas de comparación
plt.figure(figsize=(5, 10))

plt.subplot(3, 1, 1)
plt.plot(t, y1_euler, label='Método de Euler')
plt.title('Método de Euler', fontsize=16, loc='right')
plt.title('GRUPO 6', loc="left")
plt.xlabel('t')
plt.ylabel('z(t)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.grid(which='minor', axis='both', linestyle='--', linewidth=0.5)

plt.subplot(3, 1, 2)
plt.plot(t, y1_heun, label='Método de Heun', color='orange')
plt.title('Método de Heun', fontsize=16, loc='right')
plt.title('GRUPO 6', loc="left")
plt.xlabel('t')
plt.ylabel('z(t)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.grid(which='minor', axis='both', linestyle='--', linewidth=0.5)

plt.subplot(3, 1, 3)
plt.plot(t, y1_rk, label='Método de Runge-Kutta (RK4)', color='green')
plt.title('Método de Runge-Kutta (RK4)', fontsize=16, loc='right')
plt.title('GRUPO 6', loc="left")
plt.xlabel('t')
plt.ylabel('z(t)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.grid(which='minor', axis='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

## RESOLUCIÓN NUMÉRICA DE LA ECUACIÓN DIFERENCIAL DE ORDEN n CON LOS TRES MÉTODOS DE EXPOSICIÓN DE GRÁFICOS DONDE SE COMPARAN LAS 4 SOLUCIONES

# Métodos numéricos para resolver la ecuación diferencial de orden n

Consideremos la ecuación diferencial:

$$z''' - 4z'' - 75z' + 378z = 0; \quad z''(0) = 1, \quad z'(0) = 0, \quad z(0) = 0$$

Para resolver numéricamente, convertimos la ecuación de orden 3 en un sistema de ecuaciones de primer orden:

$$\begin{aligned}
y_1 &= z \\
y_2 &= z' \\
y_3 &= z''
\end{aligned}$$

Entonces, el sistema se convierte en:

$$\begin{aligned}
y_1' &= y_2 \\
y_2' &= y_3 \\
y_3' &= 4y_3 + 75y_2 - 378y_1
\end{aligned}$$

Con condiciones iniciales: $y_1(0) = 0$, $y_2(0) = 0$, $y_3(0) = 1$.

## 1. Método de Euler

El método de Euler se define como:

$$y_{n+1} = y_n + h f(t_n, y_n)$$

Para nuestro sistema:

$$\begin{aligned}
y_{1,n+1} &= y_{1,n} + h y_{2,n} \\
y_{2,n+1} &= y_{2,n} + h y_{3,n} \\
y_{3,n+1} &= y_{3,n} + h (4y_{3,n} + 75y_{2,n} - 378y_{1,n})
\end{aligned}$$

## 2. Método de Heun

El método de Heun (Euler mejorado) se define como:

$$\begin{aligned}
k_1 &= f(t_n, y_n) \\
k_2 &= f(t_n + h, y_n + h k_1) \\
y_{n+1} &= y_n + \frac{h}{2}(k_1 + k_2)
\end{aligned}$$

Para nuestro sistema:

$$\begin{aligned}
k_{1,1} &= y_{2,n} \\
k_{1,2} &= y_{3,n} \\
k_{1,3} &= 4y_{3,n} + 75y_{2,n} - 378y_{1,n} \\
k_{2,1} &= y_{2,n} + h k_{1,2} \\
k_{2,2} &= y_{3,n} + h k_{1,3} \\
k_{2,3} &= 4(y_{3,n} + h k_{1,3}) + 75(y_{2,n} + h k_{1,2}) - 378(y_{1,n} + h k_{1,1}) \\
y_{1,n+1} &= y_{1,n} + \frac{h}{2}(k_{1,1} + k_{2,1}) \\
y_{2,n+1} &= y_{2,n} + \frac{h}{2}(k_{1,2} + k_{2,2}) \\
y_{3,n+1} &= y_{3,n} + \frac{h}{2}(k_{1,3} + k_{2,3})
\end{aligned}$$

## 3. Método de Runge-Kutta de orden 4

El método RK4 se define como:

$$\begin{aligned}
k_1 &= f(t_n, y_n) \\
k_2 &= f(t_n + \frac{h}{2}, y_n + \frac{h}{2}k_1) \\
k_3 &= f(t_n + \frac{h}{2}, y_n + \frac{h}{2}k_2) \\
k_4 &= f(t_n + h, y_n + hk_3) \\
y_{n+1} &= y_n + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4)
\end{aligned}$$

Para nuestro sistema:

$$\begin{aligned}
k_{1,1} &= y_{2,n} \\
k_{1,2} &= y_{3,n} \\
k_{1,3} &= 4y_{3,n} + 75y_{2,n} - 378y_{1,n} \\
k_{2,1} &= y_{2,n} + \frac{h}{2}k_{1,2} \\
k_{2,2} &= y_{3,n} + \frac{h}{2}k_{1,3} \\
k_{2,3} &= 4(y_{3,n} + \frac{h}{2}k_{1,3}) + 75(y_{2,n} + \frac{h}{2}k_{1,2}) - 378(y_{1,n} + \frac{h}{2}k_{1,1}) \\
k_{3,1} &= y_{2,n} + \frac{h}{2}k_{2,2} \\
k_{3,2} &= y_{3,n} + \frac{h}{2}k_{2,3} \\
k_{3,3} &= 4(y_{3,n} + \frac{h}{2}k_{2,3}) + 75(y_{2,n} + \frac{h}{2}k_{2,2}) - 378(y_{1,n} + \frac{h}{2}k_{2,1}) \\
k_{4,1} &= y_{2,n} + hk_{3,2} \\
k_{4,2} &= y_{3,n} + hk_{3,3} \\
k_{4,3} &= 4(y_{3,n} + hk_{3,3}) + 75(y_{2,n} + hk_{3,2}) - 378(y_{1,n} + hk_{3,1}) \\
y_{1,n+1} &= y_{1,n} + \frac{h}{6}(k_{1,1} + 2k_{2,1} + 2k_{3,1} + k_{4,1}) \\
y_{2,n+1} &= y_{2,n} + \frac{h}{6}(k_{1,2} + 2k_{2,2} + 2k_{3,2} + k_{4,2}) \\
y_{3,n+1} &= y_{3,n} + \frac{h}{6}(k_{1,3} + 2k_{2,3} + 2k_{3,3} + k_{4,3})
\end{aligned}$$

## 4. Método de Adams-Bashforth de orden 4

El método de Adams-Bashforth de orden 4 es un método de pasos múltiples que utiliza los valores de los cuatro pasos anteriores para calcular el siguiente. La fórmula general es:

$$y_{n+1} = y_n + \frac{h}{24}(55f_n - 59f_{n-1} + 37f_{n-2} - 9f_{n-3})$$

donde $f_n = f(t_n, y_n)$.

Para nuestro sistema:

$$\begin{aligned}
y_{1,n+1} &= y_{1,n} + \frac{h}{24}(55y_{2,n} - 59y_{2,n-1} + 37y_{2,n-2} - 9y_{2,n-3}) \\
y_{2,n+1} &= y_{2,n} + \frac{h}{24}(55y_{3,n} - 59y_{3,n-1} + 37y_{3,n-2} - 9y_{3,n-3}) \\
y_{3,n+1} &= y_{3,n} + \frac{h}{24}[55(4y_{3,n} + 75y_{2,n} - 378y_{1,n}) \\
&\quad - 59(4y_{3,n-1} + 75y_{2,n-1} - 378y_{1,n-1}) \\
&\quad + 37(4y_{3,n-2} + 75y_{2,n-2} - 378y_{1,n-2}) \\
&\quad - 9(4y_{3,n-3} + 75y_{2,n-3} - 378y_{1,n-3})]
\end{aligned}$$

Nota: Para iniciar el método de Adams-Bashforth, necesitamos los primeros cuatro puntos. Estos se pueden obtener utilizando el método de Runge-Kutta de orden 4 para los primeros cuatro pasos.

#ANÁLISIS DE ERRORES DE PRIMER ORDEN PARA MÉTODOS NUMÉRICOS DE EDO

Consideremos la ecuación diferencial:

$$\frac{dy}{dx} = \frac{2xy}{3x^2 - 6y^2}, \quad y(1) = 2$$

## 1. Método de Euler

### Fórmula general:
$$y_{n+1} = y_n + h \cdot f(x_n, y_n)$$

### Error Local de Truncamiento (ELT):
- Orden $O(h^2)$
- $ELT = y(x_{n+1}) - (y_n + h \cdot f(x_n, y_n))$

### Error Global (EG):
- Orden $O(h)$

### Cálculo de iteraciones:

1. Primera Iteración $(n = 0)$:
   $$f(1, 2) = \frac{2 \cdot 1 \cdot 2}{3 \cdot 1^2 - 6 \cdot 2^2} = \frac{4}{3 - 24} = -\frac{21}{4} \approx -0.1905$$
   $$y_1 = 2 + 0.1 \cdot (-0.1905) = 1.98095$$

2. Segunda Iteración $(n = 1)$:
   $$f(1.1, 1.98095) \approx \frac{4.36009 - 19.89}{4.36009} \approx -0.2193$$
   $$y_2 = 1.98095 + 0.1 \cdot (-0.2193) \approx 1.95902$$

3. Tercera Iteración $(n = 2)$:
   $$f(1.2, 1.95902) \approx \frac{4.70165 - 18.72}{4.70165} \approx -0.2511$$
   $$y_3 = 1.95902 + 0.1 \cdot (-0.2511) = 1.93391$$

## 2. Método de Heun

### Fórmula general:
$$y_{n+1} = y_n + \frac{h}{2}[f(x_n, y_n) + f(x_n + h, y_n + h \cdot f(x_n, y_n))]$$

### Errores:
- Error Local de Truncamiento: $O(h^3)$
- Error Global: $O(h^2)$

## 3. Método de Runge-Kutta de Segundo Orden (RK2)

### Fórmulas:
$$\begin{aligned}
k_1 &= h \cdot f(x_n, y_n) \\
k_2 &= h \cdot f(x_n + \frac{h}{2}, y_n + \frac{k_1}{2}) \\
y_{n+1} &= y_n + k_2
\end{aligned}$$

### Errores:
- Error Local de Truncamiento: $O(h^3)$
- Error Global: $O(h^2)$

## Comparación de Errores

| Método | Error Global |
|--------|--------------|
| Euler  | $O(h)$       |
| Heun   | $O(h^2)$     |
| RK2    | $O(h^2)$     |

## Conclusión

Los métodos de segundo orden (Heun y RK2) ofrecen mayor precisión que el método de Euler. El error global disminuye más rápidamente al reducir el tamaño del paso $h$ en los métodos de orden superior, lo que los hace preferibles para obtener soluciones más precisas en ecuaciones diferenciales.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos la función de la ecuación diferencial
def dydx(x, y):
    return (2 * x * y) / (3 * x**2 - 6 * y**2)

# Condiciones iniciales y parámetros
x0, y0 = 1, 2
h = 0.1
n_steps = 3  # Número de pasos (cambiar según sea necesario)
x_values = [x0 + i * h for i in range(n_steps + 1)]

# Solución analítica aproximada para referencia
sol = solve_ivp(dydx, [x0, x0 + n_steps * h], [y0], method='RK45', t_eval=x_values)
y_exact = sol.y[0]

# Método de Euler
y_euler = [y0]
for i in range(n_steps):
    y_next = y_euler[-1] + h * dydx(x_values[i], y_euler[-1])
    y_euler.append(y_next)

# Método de Heun
y_heun = [y0]
for i in range(n_steps):
    fxy = dydx(x_values[i], y_heun[-1])
    y_predict = y_heun[-1] + h * fxy
    y_correct = y_heun[-1] + (h / 2) * (fxy + dydx(x_values[i + 1], y_predict))
    y_heun.append(y_correct)

# Método de Runge-Kutta de segundo orden (RK2)
y_rk2 = [y0]
for i in range(n_steps):
    k1 = h * dydx(x_values[i], y_rk2[-1])
    k2 = h * dydx(x_values[i] + h / 2, y_rk2[-1] + k1 / 2)
    y_rk2.append(y_rk2[-1] + k2)

# Cálculo de errores absolutos
error_euler = np.abs(np.array(y_euler) - np.array(y_exact))
error_heun = np.abs(np.array(y_heun) - np.array(y_exact))
error_rk2 = np.abs(np.array(y_rk2) - np.array(y_exact))

# Graficar errores
plt.figure(figsize=(10, 6))
plt.plot(x_values, error_euler, label='Error Euler', marker='o')
plt.plot(x_values, error_heun, label='Error Heun', marker='o')
plt.plot(x_values, error_rk2, label='Error Runge-Kutta (RK2)', marker='o')
plt.title('Análisis de Errores de Métodos Numéricos')
plt.xlabel('x')
plt.ylabel('Error Absoluto')
plt.yscale('log')  # Escala logarítmica para mejor visualización
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros
h = 0.01  # Tamaño de paso
t = np.arange(-0.25, 0.25, h)  # Intervalo de tiempo
n = len(t)

# Condiciones iniciales para cada método
y1_euler, y2_euler, y3_euler = [1], [0], [0]  # Método de Euler
y1_heun, y2_heun, y3_heun = [1], [0], [0]  # Método de Heun
y1_rk, y2_rk, y3_rk = [1], [0], [0]  # Método de Runge-Kutta de cuarto orden

# Solución de referencia (en este caso, se asume que es la solución de Runge-Kutta para alta precisión)
y1_exact = [1]
y2_exact = [0]
y3_exact = [0]

# Funciones de derivadas
def f1(y2):
    return y2

def f2(y3):
    return y3

def f3(y1, y2, y3):
    return -4 * y3 - 75 * y2 + 378 * y1

# Calculo la solución de referencia utilizando RK4 con un tamaño de paso más pequeño (h/10)
h_ref = h / 10
for i in range(int(n * 10) - 1):
    k1_y1 = h_ref * f1(y2_exact[-1])
    k1_y2 = h_ref * f2(y3_exact[-1])
    k1_y3 = h_ref * f3(y1_exact[-1], y2_exact[-1], y3_exact[-1])

    k2_y1 = h_ref * f1(y2_exact[-1] + k1_y2 / 2)
    k2_y2 = h_ref * f2(y3_exact[-1] + k1_y3 / 2)
    k2_y3 = h_ref * f3(y1_exact[-1] + k1_y1 / 2, y2_exact[-1] + k1_y2 / 2, y3_exact[-1] + k1_y3 / 2)

    k3_y1 = h_ref * f1(y2_exact[-1] + k2_y2 / 2)
    k3_y2 = h_ref * f2(y3_exact[-1] + k2_y3 / 2)
    k3_y3 = h_ref * f3(y1_exact[-1] + k2_y1 / 2, y2_exact[-1] + k2_y2 / 2, y3_exact[-1] + k2_y3 / 2)

    k4_y1 = h_ref * f1(y2_exact[-1] + k3_y2)
    k4_y2 = h_ref * f2(y3_exact[-1] + k3_y3)
    k4_y3 = h_ref * f3(y1_exact[-1] + k3_y1, y2_exact[-1] + k3_y2, y3_exact[-1] + k3_y3)

    y1_exact.append(y1_exact[-1] + (1 / 6) * (k1_y1 + 2 * k2_y1 + 2 * k3_y1 + k4_y1))
    y2_exact.append(y2_exact[-1] + (1 / 6) * (k1_y2 + 2 * k2_y2 + 2 * k3_y2 + k4_y2))
    y3_exact.append(y3_exact[-1] + (1 / 6) * (k1_y3 + 2 * k2_y3 + 2 * k3_y3 + k4_y3))

y1_exact = y1_exact[::10]
y2_exact = y2_exact[::10]
y3_exact = y3_exact[::10]

# Método de Euler
for i in range(n - 1):
    y1_euler.append(y1_euler[-1] + h * f1(y2_euler[-1]))
    y2_euler.append(y2_euler[-1] + h * f2(y3_euler[-1]))
    y3_euler.append(y3_euler[-1] + h * f3(y1_euler[-1], y2_euler[-1], y3_euler[-1]))

# Método de Heun
for i in range(n - 1):
    y1_predict = y1_heun[-1] + h * f1(y2_heun[-1])
    y2_predict = y2_heun[-1] + h * f2(y3_heun[-1])
    y3_predict = y3_heun[-1] + h * f3(y1_heun[-1], y2_heun[-1], y3_heun[-1])

    y1_heun.append(y1_heun[-1] + (h / 2) * (f1(y2_heun[-1]) + f1(y2_predict)))
    y2_heun.append(y2_heun[-1] + (h / 2) * (f2(y3_heun[-1]) + f2(y3_predict)))
    y3_heun.append(y3_heun[-1] + (h / 2) * (f3(y1_heun[-1], y2_heun[-1], y3_heun[-1]) + f3(y1_predict, y2_predict, y3_predict)))

# Método de Runge-Kutta de cuarto orden (RK4)
for i in range(n - 1):
    k1_y1 = h * f1(y2_rk[-1])
    k1_y2 = h * f2(y3_rk[-1])
    k1_y3 = h * f3(y1_rk[-1], y2_rk[-1], y3_rk[-1])

    k2_y1 = h * f1(y2_rk[-1] + k1_y2 / 2)
    k2_y2 = h * f2(y3_rk[-1] + k1_y3 / 2)
    k2_y3 = h * f3(y1_rk[-1] + k1_y1 / 2, y2_rk[-1] + k1_y2 / 2, y3_rk[-1] + k1_y3 / 2)

    k3_y1 = h * f1(y2_rk[-1] + k2_y2 / 2)
    k3_y2 = h * f2(y3_rk[-1] + k2_y3 / 2)
    k3_y3 = h * f3(y1_rk[-1] + k2_y1 / 2, y2_rk[-1] + k2_y2 / 2, y3_rk[-1] + k2_y3 / 2)

    k4_y1 = h * f1(y2_rk[-1] + k3_y2)
    k4_y2 = h * f2(y3_rk[-1] + k3_y3)
    k4_y3 = h * f3(y1_rk[-1] + k3_y1, y2_rk[-1] + k3_y2, y3_rk[-1] + k3_y3)

    y1_rk.append(y1_rk[-1] + (1 / 6) * (k1_y1 + 2 * k2_y1 + 2 * k3_y1 + k4_y1))
    y2_rk.append(y2_rk[-1] + (1 / 6) * (k1_y2 + 2 * k2_y2 + 2 * k3_y2 + k4_y2))
    y3_rk.append(y3_rk[-1] + (1 / 6) * (k1_y3 + 2 * k2_y3 + 2 * k3_y3 + k4_y3))

# Calcular errores absolutos
error_euler = np.abs(np.array(y1_euler) - np.array(y1_exact))
error_heun = np.abs(np.array(y1_heun) - np.array(y1_exact))
error_rk = np.abs(np.array(y1_rk) - np.array(y1_exact))

# Graficar errores
plt.figure(figsize=(10, 6))
plt.plot(t, error_euler, label='Error Euler', linestyle='--', marker='o')
plt.plot(t, error_heun, label='Error Heun', linestyle='--', marker='o')
plt.plot(t, error_rk, label='Error Runge-Kutta (RK4)', linestyle='--', marker='o')
plt.title('Comparación de Errores de Métodos Numéricos', fontsize=14)
plt.xlabel('Tiempo $t$', fontsize=12)
plt.ylabel('Error Absoluto', fontsize=12)
plt.yscale('log')  # Escala logarítmica para visualizar mejor los errores pequeños
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend()
plt.show()
```

Conclusiones
Precisión y Error: Entre los tres métodos, el método de Euler es el menos preciso debido a su error de truncamiento de orden más alto. El método de Heun mejora significativamente la precisión al incluir una corrección, pero el método de Runge-Kutta de cuarto orden (RK4) ofrece la mejor precisión con el menor error global de truncamiento.

Aplicabilidad y Coste Computacional:

Euler: Aunque es el más simple y menos costoso computacionalmente, requiere pasos de tiempo muy pequeños para problemas con alta variabilidad, lo que aumenta el costo computacional en sistemas más complejos.
Heun: Proporciona un buen equilibrio entre precisión y costo computacional, ideal para problemas que no requieren la precisión extrema de RK4.
RK4: A pesar de ser más computacionalmente intensivo, es el método de elección para sistemas donde la precisión es crucial y los recursos computacionales lo permiten.
Estabilidad y Rigidez: En problemas rígidos o con soluciones que cambian rápidamente, RK4 y Heun ofrecen una mejor estabilidad en comparación con Euler. Para problemas de alta rigidez, métodos más avanzados o variantes de RK4 podrían ser necesarios.

Recomendaciones:

Método de Euler: Utilizar solo para problemas simples o cuando los recursos computacionales son muy limitados y la precisión no es crítica.
Método de Heun: Adecuado para problemas moderadamente complejos donde se necesita un equilibrio entre precisión y costo.
Método de RK4: Ideal para problemas complejos o donde la precisión es extremadamente importante, como en simulaciones físicas precisas o modelos científicos.
En resumen, la elección del método depende de la precisión requerida, la naturaleza del problema y los recursos computacionales disponibles. En situaciones donde se requiere alta precisión y estabilidad, el método RK4 es generalmente preferido.
